# Relatório de Análise VI

```python
import pandas as pd
```

```python
dados = pd.read_csv('csvfiles/aluguel_residencial.csv', sep = ';')
```

```python
dados.head(18)
```

## Valor Bruto

```python
dados ['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados
```

## Metro Quadrado

```python
dados ['Valor Metro²'] = (dados ['Valor'] / dados ['Area']).round(2)
```

```python
dados
```

## Valor Bruto Metro Quadrado

```python
dados ['Valor Bruto Metro²'] = (dados ['Valor Bruto'] / dados ['Area']).round(2)
```

## Excluindo As Variáveis

```python
dados.drop(['Valor Bruto', 'Valor Bruto Metro²'], axis = 1, inplace = True)
```

```python
dados
```

## Tipaagem Geral dos Tipos de Residêncial 

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados.head(20)
```

```python
dados.to_csv('csvfiles/aluguel_residencial.csv', sep = ';', index = False)
```
